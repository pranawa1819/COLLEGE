##### Write a function to find the longest word in the text?

```python
def longest_word(text: str):
   text = text.replace('.','')
   splitted = text.split()
  
   
   char_count={}
   for word in splitted:
     char_count[word] = len(word)
   keys= list(char_count.keys())
   values=list(char_count.values())
   
   print(keys)
   print(values)
   
   largest =0
   for ele in values:
       if ele>largest:
           largest = ele
   idx = values.index(largest)
   print("The largest word length is : ",largest)
   print("The index of largest word is:",idx)
       
  
    
    
longest_word(
    "This is template statement for testing. "
)
```
